```python
# Importing modules.
import pandas as pd
import matplotlib.pyplot as plt

# Assigning filename.
filename = 'Assignment_dataset.xlsx'

# Reading the file into  dataframe.
df = pd.read_excel(filename)

# Column names to plot
attributes = ['Attribute1', 'Attribute2']

# Creating a box plot.
boxprops = dict(color='blue')
medianprops = dict(color='red')
whiskerprops = dict(color='green')

df[attributes].boxplot(boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops)
plt.title('Box Plot for Attribute1 and Attribute2')
plt.ylabel('Values')
plt.show()	
```

## Comment

The box plot shows that Attribute1 and Attribute2 have similar medians, but Attribute2 has a slightly wider spread in its upper values, with higher variability compared to Attribute1. There are no extreme outliers in the dataset. Also, both attributes are somewhat symmetric,which shows that the data is not heavily skewed.
